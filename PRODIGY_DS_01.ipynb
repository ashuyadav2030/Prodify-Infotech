## Data Visualization

**Total population according to Sex**

```python
sex_values = df.groupby('Sex')['Value'].sum()

# Create a bar chart
plt.figure(figsize=(8, 6))
ax = sex_values.plot(kind='bar', color=['skyblue', 'lightcoral'])
plt.title('Total Value by Sex')
plt.xlabel('Sex')
plt.ylabel('Total Value')
plt.grid(axis='y')
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'), 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha = 'center', va = 'center', 
                fontsize=12, color='black', xytext=(0, 5), 
                textcoords = 'offset points')

plt.show()
```

**Total Population By age**

```python
age_values = df.groupby('Age')['Value'].sum()

plt.figure(figsize=(20, 8))
ax = age_values.plot(kind='bar', colormap='RdBu')
plt.title('Total Value by Age')
plt.xlabel('Age')
plt.ylabel('Total Value')
plt.grid(axis='y')
plt.show()
```

**Age VS Value By Sex**

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Age', y='Value', hue='Sex', palette='coolwarm', s=100)
plt.title('Age vs. Value by Sex')
plt.xlabel('Age')
plt.ylabel('Value')
plt.grid(True)
plt.legend(title='Sex')
plt.show()
```
